```python
import numpy as np
import matplotlib.pyplot as plt
import math, cmath, colorsys
plt.rcParams['text.usetex'] = True

theta = np.arange(0, 10*math.pi, 0.001)
x_values = []
y_values = []
for t in theta:
    x_values.append(0.6*math.cos((4/5)*t)*math.cos(t))
    y_values.append(0.6*math.cos((4/5)*t)*math.sin(t))

x_values_ = np.arange(-1.0, 1.0, 0.001)
y_values_ = np.arange(-1.0, 1.0, 0.001)
z_values = []
for k in x_values_:
    for l in y_values_:
        z_values.append(complex(k, l))

z_values_0 = []
for t in z_values:
    w = cmath.polar(t)
    if math.isclose(w[0],max(abs(0.6*math.cos((4/5)*(w[1]+2*math.pi*n))) for n in range(-3,3)), rel_tol = 1e-3):
        z_values_0.append(t)

z_values_1 = np.array(z_values_0)

plt.figure()
plt.axis([-0.82, 0.82, -0.82, 0.82])
plt.title(r'$f(z)=z,~r\le \frac{3}{5} |\cos(\frac{4}{5} \theta)|$')
plt.xlabel(r'Re(z)')
plt.ylabel(r'Im(z)')
plt.grid(True)
plt.xticks(np.arange(-0.8, 0.81, 0.4))
plt.yticks(np.arange(-0.8, 0.81, 0.4))
plt.plot(x_values, y_values, color='black', linewidth=0.9)
plt.scatter(z_values_1.real, z_values_1.imag, s=3, color='black')
plt.savefig('1.png', dpi=300)
plt.show()
```

```python
z_values_1 = np.array(z_values_1)
fz_values = -2j*np.tanh(z_values_1)

x_values_fz = []
y_values_fz = []
z_values_fz =[]
for t in theta:
    x_values_fz.append(0.6*math.cos((4/5)*t)*math.cos(t))
    y_values_fz.append(0.6*math.cos((4/5)*t)*math.sin(t))
    w = complex(0.6*math.cos((4/5)*t)*math.cos(t), 0.6*math.cos((4/5)*t)*math.sin(t))
    z_values_fz.append(w)
k = -2j*np.tanh(z_values_fz)

plt.figure()
plt.axis([-1.4, 1.4, -1.1, 1.1])
plt.title(r'$f(z)=-2i\tanh(z),~r\le \frac{3}{5} |\cos(\frac{4}{5} \theta)|$')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.xticks(np.arange(-1., 1.1, 0.5))
plt.yticks(np.arange(-1., 1.1, 0.5))
plt.grid(True)
plt.plot(k.real, k.imag, color='black', linewidth=0.65)
plt.scatter(fz_values.real, fz_values.imag, s=1, color='black')
plt.savefig('2.png', dpi=300)
plt.show()
```

```python
theta = np.arange(0, 10*math.pi, 0.001)
theta1 = np.arange(0, 2*math.pi, 0.001)
theta2 = np.arange(0, math.pi, 0.001)

plt.figure()
plt.axis([-0.64, 0.64, -0.64, 0.64])
plt.title(r'$f(z)=z,~r\le \frac{3}{5} |\cos(\frac{4}{5} \theta)|$')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.xticks(np.arange(-0.64, 0.641, 0.32))
plt.yticks(np.arange(-0.64, 0.641, 0.32))

i=0.03
while i <= 0.61:
    x_values_s = []
    y_values_s = []
    for t in theta1:
        x_values_s.append(i*math.cos(t))
        y_values_s.append(i*math.sin(t))
    plt.plot(x_values_s, y_values_s, color='black', linewidth=0.2)
    i += 0.03

i=math.pi/16
while i <= math.pi/2:  
    y_values_s = []
    x_values_s = []
    for t in theta2:
        if abs(math.tan(i)*0.6*math.cos(t)) <= 0.6*math.sin(t):
            x_values_s.append(0.6*math.cos(t))
            y_values_s.append(math.tan(i)*0.6*math.cos(t))
    plt.plot(x_values_s, y_values_s, color='black', linewidth=0.2)
    y_values_s_neg = [-x for x in y_values_s]
    plt.plot(x_values_s, y_values_s_neg, color='black', linewidth=0.2)
    i += math.pi/16

plt.hlines(y=0, xmin=-0.6, xmax=0.6, color='black', linewidth=0.2)
plt.vlines(x=0, ymin=-0.6, ymax=0.6, color='black', linewidth=0.2)

plt.scatter(z_values_1.real, z_values_1.imag, s=3, color='black')
plt.savefig('3.png', dpi=300)
plt.show()
```

```python
plt.figure()
plt.axis([-1.5, 1.5, -1.2, 1.2])
plt.title(r'$f(z)=-2i\tanh(z),~r\le \frac{3}{5} |\cos(\frac{4}{5} \theta)|$')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.xticks(np.arange(-1.5, 1.51, 0.5))
plt.yticks(np.arange(-1.2, 1.21, 0.4))

i=0.03
while i <= 0.61:
    x_values_s = []
    y_values_s = []
    z_values_s = []
    for t in theta1:
        x_values_s.append(i*math.cos(t))
        y_values_s.append(i*math.sin(t))
        w = complex(i*math.cos(t), i*math.sin(t))
        z_values_s.append(w)
    k = -2j*np.tanh(z_values_s)
    plt.plot(k.real, k.imag, color='black', linewidth=0.2)
    i += 0.03

i=math.pi/16
while i <= math.pi/2:  
    y_values_s = []
    x_values_s = []
    z_values_s = []
    for t in theta2:
        if abs(math.tan(i)*0.6*math.cos(t)) <= 0.6*math.sin(t):
            x_values_s.append(0.6*math.cos(t))
            y_values_s.append(math.tan(i)*0.6*math.cos(t))
            w = complex(0.6*math.cos(t), math.tan(i)*0.6*math.cos(t))
            z_values_s.append(w)
    k = -2j*np.tanh(z_values_s)
    plt.plot(k.real, k.imag, color='black', linewidth=0.2)
    k_neg = np.array([-x for x in k])
    plt.plot(k.real, k_neg.imag, color='black', linewidth=0.2)
    i += math.pi/16

w = complex(0, 0.6)
l = -2j*cmath.tanh(w)
plt.hlines(y=0, xmin=-l.real, xmax=l.real, color='black', linewidth=0.2)

w = complex(0.6, 0)
l = -2j*cmath.tanh(w)
plt.vlines(x=0, ymin=-l.imag, ymax=l.imag, color='black', linewidth=0.2)

plt.scatter(fz_values.real, fz_values.imag, s=1, color='black')
plt.savefig('4.png', dpi=300)
plt.show()
```

```python
def hsl_to_rgb(h, s, l):
    h /= 2*math.pi
    r, g, b = colorsys.hls_to_rgb(h, l, s)
    return int(r * 255), int(g * 255), int(b * 255)


x_values_ = np.arange(-1.8, 1.8, 0.001)
y_values_ = np.arange(-1.8, 1.8, 0.001)
z_values = []
for k in x_values_:
    for l in y_values_:
        z_values.append(complex(k, l))

z_values_0 = []
rgb=[]
for t in z_values:
    w = cmath.polar(t)
    if w[0]>max(abs(0.5*math.cos((7/3)*(w[1]+2*math.pi*n))) for n in range(-3,3)) and\
        w[0]<max(abs(1.75*math.cos((3/7)*(w[1]+2*math.pi*n))) for n in range(-5,5)):
        hsl= [w[1], 1, w[0]/(w[0]+1)]
        rgb.append(hsl_to_rgb(*hsl))
        z_values_0.append(t)

z_values_1 = np.array(z_values_0)
C = np.array(rgb)

plt.figure()
plt.axis([-1.8, 1.8, -1.8, 1.8])
plt.title(r'$f(z)=z,~\frac{1}{2} |\cos(\frac{7}{3} \theta)|\le r\le \frac{7}{4} |\cos(\frac{3}{7} \theta)|$')
plt.xlabel(r'Re(z)')
plt.ylabel(r'Im(z)')
plt.grid(True, linewidth=0.2)
plt.xticks(np.arange(-1.8, 1.81, 0.4))
plt.yticks(np.arange(-1.8, 1.81, 0.4))
plt.scatter(z_values_1.real, z_values_1.imag, c=C/255.0)
plt.savefig('6.png', dpi=300)
plt.show()
```
